```python
#Part 1 - Implement Basic Data Structures using Numpy, Pandas
import numpy as np
#Lists
#Arrays
#type
list1=[1,2,3,4,5]
print(list1)
print(type(list1))
list2=[11,12,13,14,15]
print(list2)
print(type(list2))
arr1=np.array(list1)
print(arr1)
print(type(arr1))
arr2=np.array(list2)
print(arr2)
print(type(arr2))
#mathematical operations
Add=list1+list2
print("ADDITION OF LISTS:",Add)
Sub=arr1-arr2
print("SUBTRACTION OF LISTS:",Sub)
Multi=arr1*arr2
print("MULTIPLICATION OF LISTS:", Multi)
Div=arr1/arr2
print("DIVISION OF LISTS:",Div)
Power=arr1**arr2
print("POWER OF LISTS:",Power)
#tuple
tuple1=("monitor","mouse","keyboard")
print(tuple1)
print(type(tuple1))
tuple2=(100,200,300)
print(tuple2)
print(type(tuple2))
#sin
sin=np.sin(arr1)
print("SIN VALUE:",sin)
#log
log=np.log(arr1)
print("LOG VALUE:",log)
#log2
log2=np.log2(arr1)
print("LOG2 VALUE:",log2)
#Exp
exp=np.exp(arr1)
print("EXP VALUE:",exp)
```

```python
#part-2
import pandas as pd
import matplotlib.pyplot as plt

# Data
# Data
people = ['abhi', 'arun', 'sunny', 'bunny']
ages = [25, 30, 35, 40] # Removed the extra element to match the length of other lists
heights = [145, 151, 165, 173]
weights = [45, 55, 65, 75]

# Create a DataFrame
data = {
    'Name': people,
    'Age': ages,
    'Height (cm)': heights,
    'Weight (kg)': weights
}

df = pd.DataFrame(data)
# Display the DataFrame
print(df)
# Scatter Plot: Height vs Weight
plt.figure(figsize=(8, 6))
plt.scatter(df['Height (cm)'], df['Weight (kg)'], color='blue', marker='o')
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

# Bar Chart: Heights and Weights of Individuals
plt.figure(figsize=(8, 6))
bar_width = 0.4
index = range(len(df['Name']))
plt.bar(index, df['Height (cm)'], bar_width, label='Height (cm)', color='blue')
plt.bar([i + bar_width for i in index], df['Weight (kg)'], bar_width, label='Weight (kg)', color='green')
plt.title('Bar Chart of Height and Weight')
plt.xlabel('Individuals')
plt.ylabel('Measurements')
plt.xticks([i + bar_width/2 for i in index], df['Name'])
plt.legend()
plt.grid(True)
plt.show()

# Histogram: Distribution of Age, Height, and Weight
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=5, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.subplot(1, 3, 2)
plt.hist(df['Height (cm)'], bins=4, color='orange', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.subplot(1, 3, 3)
plt.hist(df['Weight (kg)'], bins=4, color='red', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
